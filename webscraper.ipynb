```python
import bs4
from urllib.request import urlopen as req
from bs4 import BeautifulSoup as soup
import csv

#Link de la pàgina on farem webscraping
url = 'https://www.newegg.com/Laptops-Notebooks/SubCategory/ID-32?Tid=6740'

#Obrim una connexió amb la pàgina web
Client = req(url)
#Descarreguem el contingut de la pàgina en una variable
pagina = Client.read()
#Tanquem connexió
Client.close()
#html parser
pagina_soup=soup(pagina,"html.parser")
#Agafa cada producte
productes = pagina_soup.findAll("div",{"class":"item-container"})

#Obrim un arxiu .csv
file = open("ordinadors.csv",'a',encoding='utf-8',newline='')
#Capçalera
header = ["Marca","Producte","PreuActual","PreuAnterior","Rebaixa","Rating","CostEnvio"]
          
write_csv = csv.DictWriter(file,fieldnames=header)
write_csv.writeheader()

#Fem un loop sobre tots els productes
for producte in productes:

    #Agafem la marca del producte
    marca_productes = producte.findAll("div",{"class":"item-info"})
    marca = marca_productes[0].div.a.img["title"]

    #Agafem el nom del producte
    name = producte.a.img["title"] 

    #Preu Actual
    actual_productes = producte.findAll("li",{"class":"price-current"})
    preuActual = actual_productes[0].strong.text
    
    #Preu anterior,rebaixa i rating  
    try:
        preuAbans = producte.find('span',class_='price-was-data').get_text()
        percentatge = producte.find('span',class_='price-save-percent').get_text()
        rating = producte.find('span',class_='item-rating-num').get_text()
    except:
        preuAbans = "NA"
        percentatge = "NA"
        rating = "NA"

    #Agafem els costos d'enviament
    costos_productes = producte.findAll("li",{"class":"price-ship"})
    costos = costos_productes[0].text.strip()  

    #Escrivim l'arxiu
    write_csv.writerow({"Marca": marca, "Producte": name, "PreuActual": preuActual, "PreuAnterior": preuAbans,"Rebaixa":percentatge,"Rating":rating,"CostEnvio":costos})

file.close() 
```
